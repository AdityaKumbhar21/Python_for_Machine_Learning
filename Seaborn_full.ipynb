```python
#importing required libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Taking dummy dataset of available with seaborn library
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# plotting relative plot
sns.relplot(data=tips, x = 'total_bill', y = 'tip', col = 'time', hue = 'smoker',style = 'smoker', size = 'size')
```

HUE: hue is nothing but a differnetiator

```python
# setting a theme for your seaborn
sns.set_theme()
```

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
print(housing)
```

```python
housing_data = pd.DataFrame(housing.data, columns = housing.feature_names)
housing_data['PRICE'] = housing.target
```

```python
housing_data.head()
```

Distribution Plot

```python
sns.displot(housing_data['PRICE'])
```

Correlation:
- Positive Correlation
- Negative Correlation

```python
correlation = housing_data.corr()
```

Heatmap

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar = True,  square = True, fmt = '.1f',cmap='Blues', annot = True,annot_kws={'size':8})
```

Where the result is moving in positive direction then it is positively correlated and where the result is moving in negative direction then it is negatively correlated
