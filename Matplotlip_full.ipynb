Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
sine_data = np.sin(x)
cos_data = np.cos(x)
```

```python
print(x)
```

```python
print(sine_data)
```

```python
print(cos_data)
```

Plotting the graph

```python
plt.plot(x, sine_data)
plt.show()
```

```python
plt.plot(x, cos_data)
plt.show()
```

Adding title and Labels to the graph

```python
plt.plot(x, sine_data)
plt.xlabel('Angle')
plt.ylabel('Sine Value')
plt.title('Sine Wave')
plt.show()
```

```python
# Adding colors to the plot

plt.plot(x, sine_data, 'r')
plt.show()
```

Plotting multiple graphs in one graph

```python
x = np.linspace(1,10,100)
plt.plot(x, np.sin(x), 'g')
plt.plot(x, np.cos(x), 'r--') # -- for dotted line
plt.show()
```

Plotting a Bar Chart

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
prog_languages = ['Javascript', 'Python', 'Java', 'C++']
users = [140,100, 70, 60]
ax.bar(prog_languages, users)
plt.xlabel('PROGRAMMING LANGAUGES')
plt.ylabel('NUMBER OF USERS')
plt.show()
```

Plotting a Pie Chart

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
prog_languages = ['Javascript', 'Python', 'Java', 'C++']
users = [140,100, 70, 60]
ax.pie(users, labels=prog_languages, autopct='%1.1f%%')  # autopct is for
# defining the digits afet decimal
plt.show()
```

Scatter Plot

```python
x = np.linspace(0,10, 30)
y = np.sin(x)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.scatter(x, y)
plt.show()
```
